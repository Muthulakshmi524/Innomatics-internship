1.Smart Light Controller

```python
class SmartLight:

    # Constructor method to initialize light name and default status
    def __init__(self, light_name):
        self.light_name = light_name   # Store the name of the light
        self.status = "OFF"            # Initially, the light is OFF

    # Method to turn the light ON
    def turn_on(self):
        self.status = "ON"             # Change status to ON

    # Method to turn the light OFF
    def turn_off(self):
        self.status = "OFF"            # Change status to OFF

    # Method to display current light status
    def display_status(self):
        print(self.light_name + " is " + self.status)  # Print light name and status


# Creating an object for Bedroom Light
light = SmartLight("Bedroom Light")

# Turning the light ON
light.turn_on()

# Displaying the current status of the light
light.display_status()
```

2.Employee ID Card System

```python
class Employee:

    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name              # Store employee name
        self.emp_id = emp_id          # Store employee ID
        self.department = department  # Store department name

    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Creating an employee object with sample input
emp1 = Employee("Rahul", "EMP102", "IT")

# Displaying the employee ID card
emp1.display_id_card()
```

3.Mobile Contact Record


```python
class Contact:

    # Constructor to initialize contact name and phone number
    def __init__(self, name, phone):
        self.name = name        # Store the contact name
        self.phone = phone      # Store the phone number

    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Creating a contact object with sample input
contact1 = Contact("Anita", "9876543210")

# Displaying the saved contact details
contact1.display_contact()
```

4.Product Price Tag Generator


```python
class Product:

    # Constructor to initialize product name and price
    def __init__(self, name, price):
        self.name = name      # Store product name
        self.price = price    # Store product price

    # Method to display formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))  # Adding ₹ symbol before price


# Creating a product object with sample input
product1 = Product("Headphones", 2499)

# Displaying the price tag
product1.display_price_tag()
```

5.Movie Rating Display System


```python
class Movie:

    # Constructor to initialize movie name and rating
    def __init__(self, name, rating):
        self.name = name        # Store movie name
        self.rating = rating    # Store movie rating

    # Method to display movie information
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")  # Display rating out of 5


# Creating a movie object with sample input
movie1 = Movie("Inception", 4.8)

# Displaying the movie details
movie1.display_movie()
```

6.Delivery Address Manager


```python
class Delivery:

    # Constructor to initialize customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name   # Store customer name
        self.address = address               # Store delivery address

    # Method to display delivery details
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating an object with sample input
order1 = Delivery("Suman", "Hyderabad")

# Displaying the delivery information
order1.display_details()
```
