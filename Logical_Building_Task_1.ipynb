1. User Login Check



```python
# This program checks whether the entered username and password are correct.

correct_username = "admin"
correct_password = "1234"

# Taking input from the user
username_input = input("Enter username: ")
password_input = input("Enter password: ")

# Checking if both username and password match
if username_input == correct_username and password_input == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer



```python
# This program counts how many students passed and failed.

# List of student marks
marks = [45, 78, 90, 33, 60]

# Variables to store pass and fail counts
pass_count = 0
fail_count = 0

# Loop through each mark
for mark in marks:
    # If marks are greater than or equal to 50 â†’ Pass
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Printing the final result
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

3. Simple Data Cleaner

```python
# This program removes extra spaces and converts names to lowercase.

# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Loop through each name
for name in names:
    # strip() removes extra spaces
    # lower() converts text to lowercase
    cleaned_name = name.strip().lower()

    # Add cleaned name to new list
    cleaned_names.append(cleaned_name)

# Print cleaned data
print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer


```python
# This program finds the length of each message and flags long messages.

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for msg in messages:
    # Find length of message
    length = len(msg)

    print("Message:", msg)
    print("Length:", length)

    # Flag messages longer than 10 characters
    if length > 10:
        print("Flag: Long message")

    print()
```

5. Error Message Detector


```python
# This program counts how many times "ERROR" appears in logs.

# System log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Variable to count errors
error_count = 0

# Loop through each log entry
for log in logs:
    # Check if the log is an ERROR
    if log == "ERROR":
        error_count += 1

# Print total error count
print("Total ERROR messages:", error_count)
```
