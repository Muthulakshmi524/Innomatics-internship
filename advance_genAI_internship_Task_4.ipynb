1.Mobile Recharge Validation System


```python
# Function to validate recharge amount
def validate_recharge(amount):

    # List of valid recharge plans
    valid_plans = [199, 299, 349, 599]

    # Check minimum recharge amount condition
    if amount < 50:
        print("Recharge amount must be at least â‚¹50.")
        return False

    # Check if entered amount is in valid plans
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False

    # If all conditions are satisfied
    return True

while True:
    try:
        # Taking user input
        amount = int(input("Enter recharge amount: â‚¹"))

        # Calling validation function
        if validate_recharge(amount):
            print("Recharge successful! âœ…")
            break  # Exit loop if valid

        else:
            print("Please try again.\n")

    except ValueError:
        # Handling non-numeric input
        print("Invalid input! Please enter a numeric value.\n")
```

2.Inventory Reorder Alert System

```python
# Function to check stock levels
def check_inventory(products):

    # Loop through dictionary items
    for product, quantity in products.items():

        if quantity < 15:
            print(product, "â†’ Reorder Alert âš ")
        else:
            print(product, "â†’ Stock OK âœ…")


# Dictionary of products and their stock quantities
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 8,
    "Salt": 18
}

# Calling the function
check_inventory(inventory)
```

3.Student Result Processing System

```python
# Function to calculate average and return result
def process_result(marks):

    total = 0

    # Loop to calculate total marks
    for mark in marks:
        total = total + mark

    # Calculate average
    average = total / len(marks)

    print("Average Marks:", average)

    # Check pass or fail condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# List of student marks
student_marks = [60, 45, 65, 55, 70]

# Calling the function
result = process_result(student_marks)

print("Result:", result)
```

4.Cab Fare Estimator with Retry Option


```python
#Function to calculate fare
def calculate_fare(distance, peak):

    base_fare = 50
    per_km_rate = 12

    # Basic fare calculation
    total_fare = base_fare + (distance * per_km_rate)

    # Apply peak hour charge
    if peak.lower() == "yes":
        total_fare = total_fare + (0.25 * total_fare)

    return total_fare

while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(distance, peak)

        print("Total Cab Fare: â‚¹", fare)

        retry = input("Do you want to calculate again? (yes/no): ")

        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator ðŸš–")
            break

    except ValueError:
        print("Invalid input! Please enter correct values.\n")
```

5.Employee Attendance Eligibility Checker

```python
# Function to check eligibility
def check_eligibility(attendance_list):

    present_days = 0

    # Loop to count number of "P"
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage, "%")

    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P","P","P"]

# Call function
result = check_eligibility(attendance)

print("Status:", result)
```

 6.Password Strength Checker


```python
# Function to check password strength
def check_password(password):

    has_digit = False
    has_special = False

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"

    # Loop through each character in password
    for ch in password:

        # Check if character is digit
        if ch.isdigit():
            has_digit = True

        # Check if character is special character
        if ch in "@#$":
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "Strong Password âœ…"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Take password input from user
password = input("Enter your password: ")

# Call function
result = check_password(password)

print(result)
```
