1.Unique Words in a Sentence


```python
# Take input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2.Highest Salary from Employee Data


```python
# Taking employee data from user
employees = {}

n = int(input("Enter number of employees: "))

for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

3.Find Maximum and Minimum Values


```python
# Take comma separated input
numbers = list(map(int, input().split(",")))

# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

# Loop to find max and min
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4. Count Products Above a Price Threshold


```python
# Take comma separated prices from user
prices = list(map(int, input().split(",")))

# Initialize counter
count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

5.Calculate Attendance Percentage


```python
# Take attendance input from user (comma separated)
attendance = input().split(",")

# Count number of Present days
present_count = attendance.count("P")

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

# Print result
print("Attendance Percentage:", percentage)
```

 6.Remove Duplicate Phone Numbers


```python
# Take phone numbers from user (comma separated)
n = list(map(int, input().split(",")))

# Convert list to set to remove duplicates
u = set(n)

# Print unique phone numbers
print("Unique phone numbers:", u)
```

 7.Count Character Frequency


```python
# Take string input
s = input()

# Create empty dictionary
d = {}

# Count each character
for c in s:
    if c in d:
        d[c] += 1
    else:
        d[c] = 1

# Print result
print(d)
```

9.Check if a Key Exists in a Dictionary

```python
# Create dictionary
d = {
    "Bharathi": 75000,
    "Divya": 68000,
    "Radhi": 72000
}

# Take key from user
k = input("Enter employee name: ")

# Check if key exists
if k in d:
    print("Employee exists")
else:
    print("Employee does not exist")
```

10.Calculate Average Marks


```python
# Take marks from user (comma separated)
m = list(map(float, input().split(",")))

# Calculate average
a = sum(m) / len(m)

# Print result
print("Average Marks:", a)
```
