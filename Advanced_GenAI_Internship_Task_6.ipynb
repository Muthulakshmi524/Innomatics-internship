1.Smart Parking Lot Management System


```python
def smart_parking_system(capacity, logs):

    current = 0      # how many vehicles are inside now
    peak = 0         # highest number of vehicles at any time

    # go through each log one by one
    for log in logs:

        if log == "IN":
            current += 1      # vehicle entered

        elif log == "OUT":
            if current > 0:
                current -= 1  # vehicle left

        # check if this is the highest count so far
        if current > peak:
            peak = current

    # print final details
    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)

    # check parking status
    if current > capacity:
        print("Alert: Parking Capacity Exceeded!")
    else:
        print("Parking Status: Available")


# sample data
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# calling the function
smart_parking_system(capacity, logs)
```

2.Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):

    base_time = distance * 5
    # assuming 5 minutes per km normally

    # add delay based on traffic
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "Low":
        base_time += 5

    # add delay based on weather
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    elif weather == "Clear":
        base_time += 0

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function Call
estimate_delivery_time(distance, traffic, weather)
```

3.Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)
    # count how many seats are booked

    occupancy = (booked_count / total_seats) * 100
    # calculate occupancy percentage

    print("Occupancy:", int(occupancy), "%")

    # check show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional show")

    elif occupancy >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Function Call
seat_occupancy_analyzer(total_seats, booked_seats)
```

4.Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):

    total = 0

    # add all CPU readings
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)
    # calculate average CPU usage

    print("Average CPU Load:", int(average), "%")

    # classify server status
    if average < 50:
        print("Server Status: Normal")

    elif average <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# Function Call
classify_server_load(cpu_readings)
```

5.Smart Classroom Resource Usage Monitor


```python
def resource_monitor(resource_usage):

    overused = []   # store overused resources

    # check each resource one by one
    for resource, hours in resource_usage.items():

        if hours > 8:
            overused.append(resource)   # add to overused list

    # display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function Call
resource_monitor(resource_usage)
```

6.Online Event Registration Capacity Controller


```python
def event_registration_controller(capacity, registrations):

    confirmed = 0      # number of confirmed seats
    waitlist = 0       # number of waitlisted users

    # process each registration
    for i in range(registrations):

        if confirmed < capacity:
            confirmed += 1      # seat confirmed
        else:
            waitlist += 1      # added to waitlist

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    # check registration status
    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

# Function Call
event_registration_controller(capacity, registrations)
```
