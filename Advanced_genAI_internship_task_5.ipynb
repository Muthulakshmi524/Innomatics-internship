1.Social Media - Post Engagement Analyzer


```python
def post_engagement(likes_list):

    # Start total likes from 0
    total_likes = 0

    # Add each like to the total
    for like in likes_list:
        total_likes += like

    # Check if the post is viral or not
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Print the final result
    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_engagement([200, 300, 150, 400])
```

3.Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_list, required_level):

    # Start total rainfall from 0
    total_rainfall = 0

    # Add each rainfall value to total
    for rain in rainfall_list:
        total_rainfall += rain

    # Calculate average rainfall
    average = total_rainfall / len(rainfall_list)

    # Check if rainfall is adequate
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Print the result
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall_checker([70, 75, 68, 80, 67], 70)
```

4.Social Media – Duplicate Account Detection


```python
def detect_duplicates(usernames):

    # Convert list to set to remove duplicates
    unique_users = set(usernames)

    # Compare lengths to check duplicates
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    # Print the result
    print("Duplicate Accounts Found:", result)

# Sample usernames
detect_duplicates(["user1", "user2", "user3", "user1"])
```

6.Agriculture – Premium Crop Price Filter


```python
def premium_crop_filter(prices):

    # Create empty list to store premium crops
    premium_crops = []

    # Check each price in the list
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    # Print the premium crop prices
    print("Premium Crops:", premium_crops)

# Sample crop prices
premium_crop_filter([1500, 2500, 1800, 3200])
```

7.System Monitoring – Application Health Checker


```python
def application_health_checker(error_count):

    # Check the error count and decide system status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Print the result
    print("Error Count:", error_count)
    print("System Status:", status)

# Sample error count to test
application_health_checker(7)
```

9.E-Learning – Student Attendance Eligibility System


```python
def attendance_checker(attendance_list):

    # Total number of classes
    total_classes = len(attendance_list)

    # Count how many classes student was present
    present_count = 0

    for status in attendance_list:
        if status == "P":
            present_count += 1

    # Calculate attendance percentage
    percentage = (present_count / total_classes) * 100

    # Check eligibility based on percentage
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    # Print the result
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", result)

# Sample attendance data
attendance_checker(["P", "P", "A", "P", "P"])
```

10.Smart Electricity Bill Analyzer


```python
def electricity_bill_analyzer(units):

    # Start bill amount from 0
    bill = 0

    # Calculate bill based on slab rules
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Decide usage category based on total bill
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

# Sample units consumed
total_bill, usage_status = electricity_bill_analyzer(250)

print("Units Consumed:", 250)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
