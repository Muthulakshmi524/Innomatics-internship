1.Employee Performance Bonus Eligibility


```python
# Employee performance dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Find the highest performance score
highest_score = max(employees.values())

#Find all employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

#Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

2.Search Query Keyword Analysis


```python
query = "Buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()

# split into words
words = query.split()

# count frequency
count = {}

for word in words:
    count[word] = count.get(word, 0) + 1

# keep only words with frequency > 1
result = {}

for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

3.Sensor Data Validation


```python
# List of sensor readings where index represents hour
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid (even) readings
valid_readings = []

# Loop through each reading using index
for index in range(len(sensor_readings)):

    # Check if the reading is even (valid)
    if sensor_readings[index] % 2 == 0:

        # Store as (hour_index, reading_value)
        valid_readings.append((index, sensor_readings[index]))

# Display the valid sensor readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4.Email Domain Usage Analysis


```python
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Extract domain and count frequency
for email in emails:

    # Split email using '@' and take domain part
    domain = email.split("@")[1]

    # Count domain occurrences
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Total number of emails
total_emails = len(emails)

# Calculate and display percentage usage
for domain in domain_count:

    # Calculate percentage
    percentage = (domain_count[domain] / total_emails) * 100

    # Display result
    print(f"{domain}: {int(percentage)}%")
```

5.Sales Spike Detection


```python
# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Calculate 30% above average threshold
threshold = average_sales * 1.3

# Check for sales spike
for i in range(len(sales)):

    # If sale is more than 30% above average
    if sales[i] > threshold:

        # Print day number (index + 1) and sale value
        print(f"Day {i+1}: {sales[i]}")
```

6.Duplicate User ID Detection


```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store count of each user ID
id_count = {}

# Count frequency of each user ID
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Display only duplicate IDs (count > 1)
for user, count in id_count.items():

    # Check if user ID appears more than once
    if count > 1:
        print(f"{user} → {count} times")
```
